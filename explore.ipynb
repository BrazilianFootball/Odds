```python
df = pd.DataFrame()
for file in glob('odds/*.csv'): df = pd.concat([df, pd.read_csv(file)])
df = df[df['Competition'].isin(['FootballBrazilSerie A', 'FootballBrazilSerie B'])]
df.sort_values(['Year', 'Odd group'], ignore_index=True, inplace=True)
df[['Home club', 'Away club']] = df['Game'].str.split(' - ', expand=True)
df['Home club'] = df['Home club'].apply(lambda x : clubs_map[x.strip()])
df['Away club'] = df['Away club'].apply(lambda x : clubs_map[x.strip()])
df
```

```python
clubs = np.unique(np.hstack([df['Home club'].values, df['Away club'].values]))
clubs = [club.replace('(playoff)', '').strip() for club in clubs]
clubs = np.unique(clubs)
clubs
```
